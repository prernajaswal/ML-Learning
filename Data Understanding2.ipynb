a) CATEGORICAL UNIVARIANT EDA

1 Count Plot

```python
#sns.countplot (df["sentiment"])
import matplotlib.pyplot as plt
#import seaborn as sns
plt.figure(figsize=(5,5))   
# adds colors types  buts its mainly set1 for the eyecatchig color & set2 is for pistel color 
#palette="deep","muted","bright","pastel","dark","colorblind"

sns.countplot(data=df, x="sentiment",hue="sentiment",palette="Set1",legend=True)

plt.xticks(rotation=45)
plt.title("Sentiment Distribution")
plt.tight_layout()
plt.show()
```

```python
df["sentiment"].value_counts().plot(kind='bar') 
```

2 Pie Chart

```python
df['sentiment'].value_counts().plot(kind="pie",autopct=lambda p: f"{p:.1f}%")
```
