```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix,ConfusionMatrixDisplay, precision_score, recall_score, f1_score
```

```python
iris=load_iris()
x=iris.data
y=iris.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=43)
model1=DecisionTreeClassifier( criterion="gini",max_depth=3, random_state=43)
```

```python
model2=DecisionTreeClassifier(criterion="entropy",random_state=43)
```

```python
model1.fit(x_train,y_train)
```

```python
model2.fit(x_train, y_train)
```

```python
y1_pred=model1.predict(x_test)
y2_pred = model2.predict(x_test)
```

```python
# print("which method is used to draw the decision treee  1=model1(gini)  2=model2(entropy)  select the choice")
choice=int(input("enter your choice"))
if choice==1:
    print("Accuracy",accuracy_score(y_test,y1_pred))
    print("precision",precision_score(y_test, y1_pred, average='macro'))
    print("recall",recall_score(y_test, y1_pred, average='macro'))
    print("f1_score ",f1_score(y_test, y1_pred, average='macro'))
    plt.figure(figsize=(15,8))  
    plot_tree( model1,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
    plt.show()  
    cm=confusion_matrix(y_test,y1_pred)
    print(cm)
    shows=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=iris.target_names)
    shows.plot(cmap="Blues") 
    plt.show()
elif choice == 2:
    print("Accuracy (Gini):", accuracy_score(y_test, y2_pred))
    print("precision",precision_score(y_test, y2_pred, average='macro'))
    print("recall",recall_score(y_test, y2_pred, average='macro'))
    print("f1_score ",f1_score(y_test, y2_pred, average='macro'))
    plt.figure(figsize=(15, 8))
    plot_tree(model2,feature_names=iris.feature_names,class_names=iris.target_names,filled=True)
    plt.title("Decision Tree using Gini (CART)")
    plt.show() 
    cm=confusion_matrix(y_test,y2_pred)
    print(cm)
    shows=ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=iris.target_names)
    shows.plot(cmap="Blues") 
    plt.show()
else:
    print("invalid input please give valid input either enter 1 or 2")

```
